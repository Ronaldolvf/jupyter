```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
import matplotlib as mpl
import datetime as dt
import matplotlib.pyplot as fig
import seaborn as sns
import bar_chart_race as bcr
```
