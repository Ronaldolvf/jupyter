```python
# Importando as bibliotecas que serão utilizadas. 
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
import datetime as dt
```

```python
# Importando os dados do aquivo "IBOVDiA3.csv".
dados = pd.read_csv('IBOVDiA3.csv')                      
display(dados)
```

```python
# Selecinando os tickers das ações. (Que se localizam na quarta coluna).
tickers = dados["CD"]

# Informando a data inicial e final.
data_inicial = dt.datetime(2022,5,30)
data_final   = dt.datetime(2022,6,3)

# Jogando os dados para um lista.
lista = [ "BBDC4.SA","BBAS3.SA", "ITUB4.SA", "SANB4.SA", "VALE3.SA", "VIIA3.SA", "YDUQ3.SA"]

# Buscando a cotação das ações (tickers) importadas via "yahoo finance,".
df = wb.DataReader (lista, "yahoo", data_inicial, data_final) 
df
```

```python
# Calculando a variação diária da ação usando a função "pct_change() ".
df['Close'] = df['Close'].pct_change()         

# Selecionando a coluna "Close" e multiplicando por 100 as cotações da linha 01.
# Colocando 2 colchetes, cria-se um data-frame. Essa função ocorre graças ao pandas. 
test = df[['Close']].round(4)*100    
test.rename_axis('dataaaaa')              
```

```python
# Buscando os dados da linha 01.
ibov = test.iloc[1].round(3)

# sorted é uma função usada para colocar em ordem crescente dos dados.
test = sorted (ibov)
test
```

```python
test = pd.DataFrame(test)
test.rename (columns= {0 : "Cotação"}, inplace = True)
test
```

```python
test.to_excel('ibov_cotação.xlsx')       # Exportando via arquivo xlsx.
```
