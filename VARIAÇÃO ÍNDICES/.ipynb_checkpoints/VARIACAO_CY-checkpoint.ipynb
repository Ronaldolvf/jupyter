```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
import matplotlib as mpl
import datetime as dt
import matplotlib.pyplot as fig
import seaborn as sns
```

```python
data_inicial = dt.datetime(2021,1,1)
data_final   = dt.datetime(2022,5,9)
lista        = ["ETH-USD", "BTC-USD"]
df           = wb.DataReader (lista, "yahoo", data_inicial, data_final) 
df["Close"].plot(subplots = True,
                 lw = 3, 
                 figsize = (20, 10)
                 )
                                           

fig.yticks (fontsize = 15)
plt.xticks (fontsize = 15)
```

```python
df['Close'] = df['Close'].pct_change()           

test = df[['Close']].round(4)*100                               
test
```
