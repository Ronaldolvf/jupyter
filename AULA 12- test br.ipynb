```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb    # Para usa-lo, precisa instala o (pip install pandas data-reader)
import matplotlib.pyplot as plt
```

```python
tickers = ['SAPR4.SA', 'COGN3.SA', 'LREN3.SA', 'AMER3.SA', 'BBSE3.SA', 'BBAS3.SA']
sec_data = pd.DataFrame()
for t in tickers:
    sec_data[t] = wb.DataReader(t, data_source = 'yahoo', start = '2022/01/01', end = "2022/05/02") ['Adj Close']  #Adj Close é a coluna
print (sec_data.head())
print (sec_data.tail())
```

```python
sec_returns = np.log(sec_data / sec_data.shift(1))  # Calcula o retorno logarítimo de ambas empresas 
sec_returns
```
