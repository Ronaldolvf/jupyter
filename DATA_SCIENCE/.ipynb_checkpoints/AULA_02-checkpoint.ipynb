# Aula_05

```python
dados = pd.read_csv ("aluguel_residencial.csv", sep = ";")
dados.head()
```

```python
# Selecioando apenas os apartamentos
selecao = dados ["Tipo"] == "Apartamento"
# Inserindo apenas os apartamento no DataFrame
n1 = dados[selecao].shape[0]
n1
```

```python
#Selecione os imóveis classificados com tipos 'Casa', 'Casa de Condomínio' e 'Casa de Vila'.
selecao = (dados['Tipo'] == 'Casa') | (dados['Tipo'] == 'Casa de Condomínio') | (dados['Tipo'] == 'Casa de Vila')
n2 = dados[selecao].shape[0]
n2
```

```python
#Selecione os imóveis com área entre 60 e 100 metros quadrados, incluindo os limites.
#60 <= Area <=100
selecao = (dados["Area"] >= 60 ) &  (dados["Area"] <= 100)
n3 = dados[selecao].shape[0]
n3
```

```python
#Selecione os imóveis que tenham pelo menos 4 quartos e aluguel menor que R$ 2.000,00.
selecao = (dados["Quartos"] < 4 ) & (dados["Valor"] < 2000.0)
n4 = dados[selecao].shape[0]
n4
```

```python
print("Nº de imóveis classificados com tipo 'Apartamento' = ({})".format(n1))
print("Nº de imóveis classificados com tipos 'Casa', 'Casa de Condomínio' e 'Casa de Vila' = ({})".format(n2))
print("Nº de imóveis com área entre 60 e 100 metros quadrados, incluindo os limites = ({})".format(n3))
print("Nº de imóveis que tenham pelo menos 4 quartos e aluguel menor que R$ 2.000,00 = ({})".format(n4))
```

# Tratamento de valores nulos

```python
dados = pd.read_csv ("aluguel_residencial.csv", sep = ";")
dados.head()
```

```python
dados.isnull()
```

```python
# Buscando apenas o NaN.
dados[dados["Valor"].isnull()]
```

```python
# Eliminando so NaN da planilha.
a = dados.shape[0]
dados.dropna(subset= ["Valor"], inplace = True)
b = dados.shape[0]
a - b
```

# Tratamento condicional

```python
dados[dados['Condominio'].isnull()].shape[0]
```

```python
# Fazendo uma selecao do apartamento e do condominio que esta coo nulo.
selecao = (dados['Tipo'] == 'Apartamento') & (dados['Condominio'].isnull())
```

```python
# Colocando o "~", ele inverte o que e verdadeiro em falso e falso em verdadeiro.
A = dados.shape[0]
dados = dados[~selecao]
B = dados.shape[0]
A - B
```

```python
# Isere o 0 nas abas nulo.
dados.fillna(0, inplace = True)
dados
```

## Criando novas variaveis

```python
dados = pd.read_csv("aluguel.csv", sep = ";")
dados.head()
```

```python
# Inserindo uma nova coluna no DataFrame.
dados['Valor Bruto'] = dados['Valor'] + dados['Condominio'] + dados['IPTU']
dados.head()
```

```python
dados['Valor m2'] = (dados['Valor'] /dados['Area']).round(2)
dados.head()
```

```python
dados['Valor Bruto m2'] = (dados['Valor Bruto']/dados['Area']).round(2)
dados.head()
```

```python
casa = ['Casa', 'Casa de Condomínio', 'Casa de Vila']
```

```python
# Criando uma nova coluna onde, a lista casa atribuida a casa e ou apartamento.
dados['Tipo Agregado'] = dados['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento')
dados
```
