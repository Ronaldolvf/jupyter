# Aula_04

```python
dados = pd.read_csv("aluguel.csv", sep = ";")
dados.head()
```

```python
list (dados["Tipo"].drop_duplicates())
```

```python
residencial = ['Quitinete',
              'Casa',
              'Apartamento',
              'Casa de Condomínio',
              'Casa de Vila']
residencial
```

```python
# Com a funcao "isin" ele vai bater as informacoes da lista com o DataFrame e retorna True or False.
selecao = dados ["Tipo"].isin(residencial)
```

```python
dados_residencial = dados[selecao]
dados_residencial
```

```python
# Verificando os dados na coluna "Tipo".
list (dados_residencial["Tipo"].drop_duplicates())
```
