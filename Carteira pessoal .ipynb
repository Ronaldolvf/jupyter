```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
```

Reorno de uam carteira de ações 

```python
tickers = ['COGN3.SA', 'LREN3.SA', 'AMER3.SA', 'BBSE3.SA', '^BVSP']
mydata = pd.DataFrame()
for t in tickers:
    mydata[t] = wb.DataReader(t, data_source = 'yahoo', start = '2022') ['Adj Close']  #Adj Close é a coluna
mydata
```

```python
mydata.iloc[0]
```

```python
(mydata / mydata.iloc[0] * 100).plot(figsize = (25,13),
                                     ls = "--",          # Formarto da linha. 
                                     lw = "2",           # Tamanho da linha. 
                                     marker = "o",       # Para marca O ponto
    
                                    )       
plt.legend(loc = "lower left", fontsize = "12" )                          # Para desloca para a esquerda. 
plt.title ("Cotação bancos 2022", color = "black", fontsize = "30")        # Titulo da planilha.
plt.ylabel ("Cotação", fontsize = "22")                                    # Nome lateral.                         
plt.xlabel("Data", fontsize = "22" )                                       # Nome inferior. 




plt.show()
```

```python
mydata.loc['2021-01-20']  # Para aplicar o loc, tem que indicar que o rótulo coresponde a primeira linha. As datas são rótulos.
```

```python
mydata.iloc[0]  # Para o iloc, tem que indicar a posição de interesse. 0 é a primeira linha 
```

```python
returns = (mydata / mydata.shift(1)) -1
returns.head()
```

```python

weights = np.array([0.25, 0.25, 0.25, 0.25])   #Refere-se ao peso da ação
```

```python
np.dot(returns,weights)   #Multiplica o peso de cada ação e seu respectivo retorno
```

```python
annual_returns = returns.mean() * 168   #seus respctivos retornos 
annual_returns
```

```python
np.dot(annual_returns, weights)
```

```python
pfolio_1 = str(round(np.dot(annual_returns, weights), 4 )*100) + '%'    #RENTBILIDADE DA CERTEIRA
print (pfolio_1)
```
