b. Extract data from IEX for Ford from the 1st of January 2015. The ticker you need is ‘F’. 

Buscando os tickers das ações 

```python
tickers = ['AAPL', 'AMZN', 'VALE', 'DIS', 'BTC-USD']
new_data = pd.DataFrame()
for t in tickers:
    new_data[t] = wb.DataReader(t, data_source = 'yahoo', start = '2021-08-16') ['Adj Close']  #Adj Close é a coluna
new_data.tail()
```
