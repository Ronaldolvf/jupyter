```python
import numpy as np
import pandas as pd

from scipy import stats 
import statsmodels.api as sm

import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('Housing.xlsx')   #Importando uma tabela em excel
data
```

```python
data[['House Price', 'House Size (sq.ft.)']]
```

```python
x = data['House Size (sq.ft.)']
y = data['House Price']
```

```python
x
```

```python
y
sum(y)
```

```python
plt.scatter(x, y)
plt.show()
```

```python
plt.scatter(x, y)
plt.axis([0, 2500, 0, 1500000])   #0, 2500 e 0, 1500000 serão os valores extremos
plt.show()
```

```python
plt.scatter(x, y)
plt.axis([0, 2500, 0, 1500000])  
plt.ylabel('House Price')    #Inflorma o título do gráfico 
plt.xlabel('House Size (sq.ft')  #Inflorma o título do gráfico 
```

```python
x1 = sm.add_constant(x)  #

reg = sm.OLS(y, x1).fit()
reg.summary()
```
