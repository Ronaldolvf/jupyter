```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
```

```python
AAPL = wb.DataReader('AAPL', data_source= 'yahoo', start = '2022')
AAPL.head()  
```

```python
AAPL.tail()
```

```python
AAPL['simple_return'] = (AAPL['Adj Close'] / AAPL['Adj Close'].shift(1)) - 1  #Calculando o retorno simples do ativo. # shift(1) indica qauntos dias anterior.
print (AAPL['simple_return'])
```

```python
AAPL['simple_return'].plot(figsize = (8, 5))
plt.show()
```

```python
avg_returns_d = AAPL['simple_return']. mean()  #taxa de retorno
avg_returns_d
```

```python
avg_returns_a = AAPL['simple_return']. mean() *42 #taxa de retorno
avg_returns_a
```

```python
print (str(round(avg_returns_a, 5)* 100) + '%')
```

LOG RETURN

```python
AAPL.head()
```

```python
AAPL['log_return'] = np.log(AAPL['Adj Close'] / AAPL['Adj Close'].shift(1))  #Calculando o retorno simples do ativo. # shift(1) indica qauntos dias anterior.
print (AAPL['log_return'])
```

```python
AAPL['log_return'].plot(figsize = (8, 5))
plt.show()
```

```python
log_returns_a = AAPL['log_return']. mean() *168 #taxa de retorno
log_returns_a
```

```python
print (str(round(log_returns_a, 5)* 100) + '%')
```

Reorno de uam carteira de aes 

```python
tickers = ['BBDC4.SA', 'ITUB4.SA', 'BRSR6.SA', 'BBAS3.SA']
mydata = pd.DataFrame()
for t in tickers:
    mydata[t] = wb.DataReader(t, data_source = 'yahoo', start = '2022') ['Adj Close']  #Adj Close  a coluna
mydata
```

```python
mydata.iloc[0]
```

```python
(mydata / mydata.iloc[0] * 100).plot(figsize = (15,5));
plt.show()
```

```python
mydata.loc['2022-01-03']  # Para aplicar o loc, tem que indicar que o rtulo coresponde a primeira linha. As datas so rtulos.
```

```python
mydata.iloc[0]  # Para o iloc, tem que indicar a posio de interesse. 0  a primeira linha 
```

```python
returns = (mydata / mydata.shift(1)) -1
returns.head()
```

```python

weights = np.array([0.25, 0.25, 0.25, 0.25])   #Refere-se ao peso da ao
```

```python
np.dot(returns,weights)   #Multiplica o peso de cada ao e seu respectivo retorno
```

```python
annual_returns = returns.mean() * 168   #seus respctivos retornos 
annual_returns
```

```python
np.dot(annual_returns, weights)
```

```python
pfolio_1 = str(round(np.dot(annual_returns, weights), 4 )*100) + '%'    #RENTBILIDADE DA CERTEIRA
print (pfolio_1)
```

CALCULAR NDICES DE AES

```python
tickers = ['^DJI', '^GSPC', '^IXIC', '^BVSP']

ind_data = pd.DataFrame()

for t in tickers:
    ind_data[t] = wb.DataReader(t, data_source='yahoo', start='2021', end="2022/05/02")['Adj Close']
```

```python
ind_data.tail()
```

```python
(ind_data / ind_data.iloc[0] * 100).plot(figsize = (20,10));  #Coloca-lo para base 100
plt.show()
```

```python

ind_returns = (ind_data / ind_data.shift(1)) - 1

ind_returns
```

```python
annual_ind_returns = ind_returns.mean() * 168
annual_ind_returns
```
