```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
```

```python
tickers = ['AAPL', 'MSFT']
sec_data = pd.DataFrame()
for t in tickers:
    sec_data[t] = wb.DataReader(t, data_source = 'yahoo', start = '2022') ['Adj Close']  #Adj Close é a coluna
sec_data
```

```python
sec_returns = np.log(sec_data / sec_data.shift(1))*100  # Calcula o retorno logarítimo de ambas empresas 
sec_returns
```

Calculando o risco do portifólio 
    
    

```python
weights = np.array([0.5, 0.5])  
```

```python
pfolio_var = np.dot(weights.T, np.dot(sec_returns.cov() * 49, weights))   #Variancia da carteira anterior 
pfolio_var
```

```python
pfolio_vol = (np.dot(weights.T, np.dot(sec_returns.cov() * 49, weights))) **0.5   # Calculando a volatividade 
pfolio_vol
```

```python
print (str(round(pfolio_vol, 5) * 100) + '%')
```
