Calculando o Beta de uma ação 

```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb

tickers =['PG', '^GSPC']
data = pd.DataFrame()
for t in tickers:
    data[t] = wb.DataReader (t, data_source = 'yahoo', start ='2012-1-1', end = '2016-12-31') ['Adj Close']
```

```python
sec_returns = np.log(data / data.shift(1))  
sec_returns
```

```python
cov = sec_returns.cov() *250      # Covariancia 
cov
```

```python
cov_with_market = cov.iloc[0,1]   # Covariancia (Esse no caso é o numerador ) 
cov_with_market
```

```python
market_var = sec_returns ['^GSPC'].var() *250   # Calcular a variancia do mercado (Denominador)
market_var
```

Beta:

```python
PG_beta = cov_with_market / market_var
PG_beta
```

Calculando o CAPM

```python
PG_er = 0.025 + PG_beta * 0.05    #0.025 (Taxa livre de risco)    0.05 (É o prémio de risco )
PG_er
```

Calculando o índice Sharpe

```python
Sharpe = (PG_er - 0.025) / (sec_returns['PG'].std() *250 ** 0.5)      #
Sharpe
```
